```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Basic Plots

```python
# create x-coordinates for plot
x = np.array([2023, 2024, 2025, 2026])
# create y-coordinates for plot
y = np.array([15, 25, 30, 45])

# this creates a basic plot
plt.plot(x,y) 
# display the plot
plt.show()  
```

```python
"""
creates a plot with more specifications
marker is the shape of each point
markersize is the size of each point
markerfacecolor is the color of each point
markeredgecolor is the color of the edge of each point
linestyle styles the line
linewidth is the width of the line
color is the color of the line
"""
plt.plot(x,y, marker=".", 
              markersize=20, 
              markerfacecolor="red", 
              markeredgecolor="green", 
              linestyle="dotted", 
              linewidth=4, 
              color="yellow")
"""
the line above can also be written in shorthand like this
plt.plot(x,y, marker=".", 
              ms=20, 
              mfc="red", 
              mec="green", 
              ls="dotted", 
              lw=4, 
              color="yellow")
"""
plt.show()
```

```python
# you can also create a dictionary to pass to the plt.plot() function
line_style = dict(marker=".", 
                  ms=20, 
                  mfc="red", 
                  mec="green",
                  ls="dotted", 
                  lw=4, 
                  color="yellow")

# the ** before line_style is necessary to unpack the dictionary into separate keyword arguments
plt.plot(x,y, **line_style)
plt.show()
```

```python
# create a title 
plt.title("My chart", fontsize=20, 
                      family="Arial", 
                      fontweight="bold", 
                      color="red")

# name the x-axis
plt.xlabel("Year", fontsize=20, 
                   family="Arial", 
                   fontweight="bold", 
                   color="red")

# name the y-axis
plt.ylabel("y-axis", fontsize=20, 
                     family="Arial", 
                     fontweight="bold", 
                     color="red")

# unlike the charts above, this will only display values from the x variable along the x-axis
plt.xticks(x)
# with plt.ticks_params() you can style the ticks


"""
All of the details above are not strictly necessary, you could also do this
plt.title("My chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
"""

plt.plot(x,y, **line_style)
plt.show()
```

```python
plt.plot(x,y)

"""
you can add gride lines with the code below
plt.grid(axis="x") will only create grid lines from the x-axis
plt.grid(axis="y") will only create grid lines from the y-axis
plt.grid() creates both sets of grid lines
you can also add keyword agruments to style the grid lines just like above with line_style
"""
plt.grid() 
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum(), color="red", label="with cusum()")
ax.plot(np.random.standard_normal(1000), color="blue", label="no cumsum()")
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(["one", "two", "three", "four", "five"],rotation=45, fontsize=12)
ax.set_xlabel("First")
ax.set_ylabel("Second")
plt.title("My random chart")
# could also write the previous 3 lines in one line
# ax.set(title="My random chart", xlabel="First", ylabel="Second")
ax.legend()
plt.show()
```

# Bar Charts

```python
categories = np.array(["cats", "dogs", "birds"])
amounts = np.array([3, 4, 2])
plt.bar(categories,amounts)
plt.title("Pets")
plt.xlabel("Type of pet")
plt.ylabel("Number of pets")
plt.show()
```

<h1 id="pie-charts">Pie Charts</h1>

```python
positions = np.array(["cooks", "servers", "bartenders", "dishwashers"])
amounts = np.array([6, 8, 3, 4])
colors = np.array(["red","lightgreen","lightblue","pink"])

plt.pie(amounts, labels=positions,
                 autopct="%.2f%%", 
                 colors=colors, 
                 explode=[0,0,0,0.1], # allows a part of the pie to explode out
                 shadow=True, 
                 startangle=45) # alters the angle of the chart
plt.title("Restaurant employees")
plt.show()
```

# Histograms

```python
"""
loc is the median
scale is the standard deviation
"""

scores = np.random.normal(loc=80, scale=10, size=100)

# this clips the scores so none are below zero or above 100
scores = np.clip(scores, 0, 100)
plt.hist(scores, bins=10, 
                 color="red", 
                 edgecolor="black")
plt.title("Exam scores")
plt.xlabel("Score")
plt.ylabel("Number of students")
plt.show()
```

# Sub Plots

Figure = entire canvas

Ax = a sub plot within the canvas

While below I am using plots, you can use bar/pie/histogram instead, each one can be separate


```python
x = np.array([1,2,3,4,5])
fig, axes = plt.subplots(2,2)

axes[0,0].plot(x, x*2, color="red")
axes[0,0].set_title("First")

axes[0,1].plot(x, x)
axes[0,1].set_title("Second")

axes[1,0].plot(x, x**2)
axes[1,0].set_title("Third")

axes[1,1].plot(x, x**3)
axes[1,1].set_title("Fourth")

plt.tight_layout() # spaces all of the sub-plots out
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
data = np.random.standard_normal(30).cumsum()

ax.plot(data, color="red", linestyle="dashed", label="Default")
ax.plot(data, color="black", linestyle="dashed", drawstyle="steps-post", label="steps-post")
ax.legend()
plt.show()
```

# Pandas with matplotlib

```python
df = pd.read_csv("data.csv")
type_count = df["Type1"].value_counts(ascending=True)
# bar() vs barh(), barh() just switches the x-axis and y-axis
plt.barh(type_count.index, type_count.values, color="lightgreen", 
                                              edgecolor="black")

plt.title("Type1 Counts")
plt.xlabel("Counts")
plt.ylabel("Type")
plt.tight_layout()
plt.show()
```
