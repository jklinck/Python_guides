```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"num":[6,7,8], "color":["red","green","blue"]}, index=['a','b','c'])
df
```

```python
# remove the row where num == 6
df.drop(df[df["num"] == 6].index, axis=0, inplace=True)
df
```

```python
# drops the column name "color"
df.drop("color", axis=1, inplace=True)
df
```
