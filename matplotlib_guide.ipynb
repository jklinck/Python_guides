```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Basic plots

```python
# create x-coordinates for plot
x = np.array([2023, 2024, 2025, 2026])
# create y-coordinates for plot
y = np.array([15, 25, 30, 45])

# this creates a basic plot
plt.plot(x,y) 
# display the plot
plt.show()  
```

```python
"""
creates a plot with more specifications
marker is the shape of each point
markersize is the size of each point
markerfacecolor is the color of each point
markeredgecolor is the color of the edge of each point
linestyle styles the line
linewidth is the width of the line
color is the color of the line
"""
plt.plot(x,y, marker=".", 
              markersize=20, 
              markerfacecolor="red", 
              markeredgecolor="green", 
              linestyle="dotted", 
              linewidth=4, 
              color="yellow")
"""
the line above can also be written in shorthand like this
plt.plot(x,y, marker=".", 
              ms=20, 
              mfc="red", 
              mec="green", 
              ls="dotted", 
              lw=4, 
              color="yellow")
"""
plt.show()
```

```python
# you can also create a dictionary to pass to the plt.plot() function
line_style = dict(marker=".", 
                  ms=20, 
                  mfc="red", 
                  mec="green",
                  ls="dotted", 
                  lw=4, 
                  color="yellow")

# the ** before line_style is necessary to unpack the dictionary into separate keyword arguments
plt.plot(x,y, **line_style)
plt.show()
```

```python
# create a title 
plt.title("My chart", fontsize=20, 
                      family="Arial", 
                      fontweight="bold", 
                      color="red")

# name the x-axis
plt.xlabel("Year", fontsize=20, 
                   family="Arial", 
                   fontweight="bold", 
                   color="red")

# name the y-axis
plt.ylabel("y-axis", fontsize=20, 
                     family="Arial", 
                     fontweight="bold", 
                     color="red")

# unlike the charts above, this will only display values from the x variable along the x-axis
plt.xticks(x)
# with plt.ticks_params() you can style the ticks


"""
All of the details above are not strictly necessary, you could also do this
plt.title("My chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
"""

plt.plot(x,y, **line_style)
plt.show()
```

```python
plt.plot(x,y)

"""
you can add gride lines with the code below
plt.grid(axis="x") will only create grid lines from the x-axis
plt.grid(axis="y") will only create grid lines from the y-axis
plt.grid() creates both sets of grid lines
you can also add keyword agruments to style the grid lines just like above with line_style
"""
plt.grid() 
plt.show()
```

```python
categories = np.array(["cats", "dogs", "birds"])
amounts = np.array([3, 4, 2])
plt.bar(categories,amounts)
plt.title("Pets")
plt.xlabel("Type of pet")
plt.ylabel("Number of pets")
plt.show()
```

```python
positions = np.array(["Cooks", "Servers", "Bartenders", "Dishwashers"])
counts = np.array([7, 5, 3, 4])
colors = np.array(["blue", "orange", "red", "green"])

"""
autopct below sets the percentage and the two %% at the end adds the % after the number 
you don't have to set colors below, it will automatically pick some
explode allows you to explode one of the categories out from the chart 
    you need to pass an array with the same length as the labels argument (in this case labels=positions)
shadow adds a small shadow to each pie slice
startangle will change how the pie is positioned
"""

plt.pie(counts, labels=positions, 
                autopct="%.2f%%", 
                colors=colors, 
                explode=[0, 0, 0, 0.2], 
                shadow=True, 
                startangle=90)
plt.title("Restaurant employees")
plt.show()
```

```python
"""
Scatter plots, plt.scatter() and regular plots, plt.plot() are similar, but the difference lies in the 
fact that with scatter plots the points do not have to be in order and it doesn't draw a connecting line 
like plt.plot() does

plt.scatter() exists to show correlation and it allows for different sizes and colors for each point
"""

# notice the arrays below are not in order
hours_studied1 = np.array([4,7,6,5])
grades1 = np.array([75,95,85,80])

plt.scatter(hours_studied1, grades1, color="red", 
                                     alpha=0.5, # sets the transparency level
                                     s=200, # sets the point size
                                     label="Student 1") # sets the label for the legend

hours_studied2 = np.array([4,5,6,7])
grades2 = np.array([85,90,95,100])
                                       
plt.scatter(hours_studied2, grades2, color="blue", 
                                     alpha=0.5, # sets the transparency level
                                     s=200, # sets the point size
                                     label="Student 2") # sets the label for the legend
                                       
plt.title("Hours studied correlation with grades")
plt.xlabel("Hours studied")
plt.ylabel("Grades")
plt.legend()
plt.show()
```
